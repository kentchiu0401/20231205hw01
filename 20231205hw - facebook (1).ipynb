```python
# 請用上週學的方法，讓程式自動登入mbasic.facebook.com

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
```

```python
username = input('請輸入您的使用者名稱/帳號：')
password = input('請輸入您的密碼：')
kword = input('請輸入你要找尋的關鍵詞：')
```

```python
path = r'C:\Users\Evan\Downloads\anaconda\chromedriver-win64\chromedriver.exe'
url = 'https://mbasic.facebook.com/'
service=Service(executable_path=path)
service
driver=webdriver.Chrome(service=service)
driver.get(url)
time.sleep(5)
un_xpath = '/html/body/div/div/div[2]/div/table/tbody/tr/td/div[2]/div/div[2]/form/ul/li[1]/input'
# 讓程式根據xpath找到輸入框
element = driver.find_element(By.XPATH, un_xpath)
element
```

```python
# 讓程式在輸入框自動輸入你的帳號

element.send_keys(username)
```

```python
# 讓程式在輸入框自動輸入你的密碼

pw_element.send_keys(password)
```

```python
# 按下登入按鈕
input_xpath = '/html/body/div/div/div[2]/div/table/tbody/tr/td/div[2]/div/div[2]/form/ul/li[3]/input'
input_button =driver.find_element(By.XPATH, input_xpath)
input_button.click()
```

```python
p1 = driver.page_source
```

```python
from bs4 import BeautifulSoup

p1_soup = BeautifulSoup(p1, 'lxml')
p1_soup
```

```python

authors=[]
for i in range(0, 99):
    try:
        p = driver.page_source
        soup = BeautifulSoup(p, 'lxml')
        p_authors = soup.find_all('h3')
        for aut in p_authors:
            if aut.find('strong')!=None:
                authors.append(aut.text.strip())
        time.sleep(3)
        driver.get(soup.find('div', id="see_more_pager").find('a').get('href'))
    except:
        break

authors
```

```python
# 計算作者發文數

aut_count={}

for aut in authors:
    if aut not in aut_count:
        aut_count[aut]=1
    else:
        aut_count[aut]+=1
        
aut_count
```

```python
max_value = max(aut_count.values())
for key, value in aut_count.items():
    if value == max_value:
        print(f'發文數最多的是{key}，共計{aut_count[key]}篇。')
```
