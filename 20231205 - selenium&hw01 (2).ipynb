```python
# 一步一步來

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import requests
from tools import get_soup
import pandas as pd
from bs4 import BeautifulSoup
import time
path = r'C:\Users\Evan\Downloads\anaconda\chromedriver-win64\chromedriver.exe'
url = 'https://mbasic.facebook.com/'
service=Service(executable_path=path)
service
driver=webdriver.Chrome(service=service)
driver.get(url)
time.sleep(5)
un_xpath = '/html/body/div/div/div[2]/div/table/tbody/tr/td/div[2]/div/div[2]/form/ul/li[1]/input'
# 讓程式根據xpath找到輸入框
element = driver.find_element(By.XPATH, un_xpath)
element
```

```python
# 讓程式在輸入框自動輸入帳號
username = input("請輸入您的用戶名稱: ")
if username.isalpha() and any(ch > '\u007F' for ch in username):
    # 如果輸入包含中文字，報錯
    raise ValueError("請勿輸入中文字。")
else:
    element.send_keys(username)
```

```python
# 讓程式定位密碼輸入框

pw_xpath='/html/body/div/div/div[2]/div/table/tbody/tr/td/div[2]/div/div[2]/form/ul/li[2]/section/input'

pw_element = driver.find_element(By.XPATH, pw_xpath)
pw_element
```

```python
# 讓程式在輸入框自動輸入密碼
password = input("請輸入您的密碼: ")
if password.isalpha() and any(ch > '\u007F' for ch in password):
    # 如果輸入包含中文字，報錯
    raise ValueError("請勿輸入中文字。")
else:
    pw_element.send_keys(password)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
search_results_page_source = driver.page_source
# 使用BeautifulSoup解析HTML頁面
soup = BeautifulSoup(search_results_page_source, 'html.parser')
posters = soup.find_all('strong', class_='bh')
for poster in posters:
    poster.decompose()  # 移除包含"社團"等非人名的部分

# 找到所有剩餘的<strong>標籤並列出
remaining_posters = soup.find_all('strong')
for poster in remaining_posters:
    print(poster.get_text())

for i in range(1,100):
    try:
        # 找到“查看更多結果”按紐並點擊
        view_more_button = driver.find_element(By.XPATH, "//span[text()='查看更多結果']")
        ActionChains(driver).move_to_element(view_more_button).click().perform()
        # 等待一段時間，確保新的結果加載完成
        driver.implicitly_wait(7)
        search_results_page_source = driver.page_source

        # 使用BeautifulSoup解析HTML頁面
        soup = BeautifulSoup(search_results_page_source, 'html.parser')

        posters = soup.find_all('strong', class_='bh')
        for poster in posters:
            poster.decompose()  #  移除包含"社團"等非人名的部分

        # 找到所有剩餘的<strong>標籤並列出
        remaining_posters = soup.find_all('strong')
        for poster in remaining_posters:
            print(poster.get_text())
            
    except Exception as e:
        #撈完了就輸出訊息告知使用者
        print(f"沒有更多結果了: {e}")
        break
```
